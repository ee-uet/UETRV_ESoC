```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
from matplotlib.pyplot import *
from numpy import *
# if using a jupyter notebook
%matplotlib inline
```

```python
# spiral
n = 100000
q = 500
m = 20
r = linspace(0,m,n)
t = linspace(0,1260,n)
x = r*cos(radians(t))*q/m + q
y = r*sin(radians(t))*q/m + q
print(min(x))
print(max(x))
print(min(y))
print(max(y))
plot(x,y)
```

```python
# sine
n = 1000
x = np.arange(0, n, 1)
x_pi = x/(n/4/np.pi)
y_pi = np.sin(x_pi)
y = y_pi*500 + 500
# plt.plot(x,y)
```

```python
w = 1000
h = w
data = np.zeros((h, w, 3), dtype=np.uint8)
for p in range(0,w):
    for q in range(0,h):
        data[p,q] = [255,255,255]
old_ly = 500
for k in range(0,n):
    lx = int(x[k]);
    ly = int(y[k]);
    if ly == 1000:
        ly = ly - 1
    dly = ly - old_ly
    while dly > 0:
        data[lx, ly - dly] = [0,0,0]
        dly = dly - 1
    while dly < 0:
        data[lx, ly - dly] = [0,0,0]
        dly = dly + 1
    data[lx,ly] = [0,0,0]
    old_ly = ly
img = Image.fromarray(data, 'RGB')
img.save('prog_spiral.png')
img.show()
```
